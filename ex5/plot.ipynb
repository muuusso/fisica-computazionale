```python
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['figure.figsize'] = (12,9)
pp = PdfPages('multipage.pdf')
```

```python
N = 125
V = np.linspace(100, 200)

df = pd.read_csv('argon.csv')

plt.errorbar(df.V[-10:], df.P[-10:], yerr=df.deltaP[-10:], fmt='.', label='Monte Carlo')
plt.plot(V, N/V, label='Ideal Gas')
plt.xlabel('V', fontsize=18)
plt.ylabel('P', fontsize=18, rotation=0, labelpad=20)
plt.legend()
plt.grid()
pp.savefig()
plt.show()

V = np.linspace(200, 2000)

plt.errorbar(df.V[1:10], df.P[1:10], yerr=df.deltaP[1:10], fmt='.', label='Monte Carlo')
plt.plot(V, N/V, label='Ideal Gas')
plt.xlabel('V', fontsize=18)
plt.ylabel('P', fontsize=18, rotation=0, labelpad=20)
plt.legend()
plt.grid()
pp.savefig()
plt.show()
```

```python
plt.errorbar(df.rho, df.Cv, yerr=df.deltaCv, fmt='.')
plt.xlabel('$\\rho$', fontsize=18)
plt.ylabel('$\\frac{C_v}{N k_B}$', fontsize=18, rotation=0, labelpad=20)
plt.grid()
pp.savefig()
plt.show()
```

```python
pp.close()
```
