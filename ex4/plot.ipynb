```python
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib.backends.backend_pdf import PdfPages

plt.rcParams['figure.figsize'] = (12,9)
pp = PdfPages('multipage.pdf')
```

```python
energy = pd.read_csv('xenon_energy.csv')

plt.plot(energy.V, label='potential')
plt.plot(energy.K, label='kinetics')
plt.plot(energy.E, label='total energy')
plt.xlabel('steps', fontsize=14)
plt.ylabel('E', fontsize=14)
plt.grid()
plt.legend(loc='center right', fontsize=12)
pp.savefig()

plt.show()

plt.plot(energy.E, c='g')
plt.xlabel('steps', fontsize=14)
plt.ylabel('E', fontsize=14)
plt.grid()
pp.savefig()

plt.show()
```

```python
cvv = pd.read_csv('xenon_cvv.csv')

plt.plot(cvv.index, cvv.cvv)
plt.xlabel('steps', fontsize=14)
plt.ylabel('CVV', fontsize=14)
plt.grid()
pp.savefig()

plt.show()
```

```python
pp.close()
```
