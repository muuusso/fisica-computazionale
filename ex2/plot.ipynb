```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.constants import *

plt.rcParams['figure.figsize'] = (8,6)

from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('multipage.pdf')
```

```python
for n in np.arange(1.5,3.25,0.25):
    df = pd.read_csv(f'ns/{n:.6f}.csv')
    plt.plot(df.xi, df.theta, label=n)
    
plt.xlabel("$\\xi$", fontsize=14)
plt.ylabel("$\\tilde{{\\theta}}(\\xi)$", fontsize=14)
plt.grid()
plt.legend()
pp.savefig()
plt.show()

for n in [1,4.5]:
    df = pd.read_csv(f'ns/{n:.6f}.csv')
    plt.plot(df.xi, df.theta, label=n)
    
plt.xlabel("$\\xi$", fontsize=14)
plt.ylabel("$\\tilde{{\\theta}}(\\xi)$", fontsize=14)
plt.grid()
plt.legend()
pp.savefig()
plt.show()
```
