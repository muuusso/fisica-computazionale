```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,6)
import numpy as np
```

```python
k = 1
N = 100
h = 4*np.pi / N

y = [0]
v = [1]
for i in range(N):
    v.append(v[i] - h*k*y[i])
    y.append(y[i] + h*v[i+1])
    
t = np.linspace(0,4*np.pi, 200)
t_ = [h*i for i in range(N+1)]
plt.plot(t, np.sin(k**.5 *t), 'r')
plt.scatter(t_, y)
plt.grid()
plt.show()
```

```python
plt.scatter(t_, abs(y-np.sin(t_)))
plt.grid()
plt.show()
```
