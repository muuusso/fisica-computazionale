```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('multipage.pdf')

plt.rcParams['figure.figsize'] = (10, 8)
```

```python
x = np.linspace(.99, 2.5, 1000)
V = lambda x: 4*(x**(-12) - x**(-6))
```

```python
H2 = pd.read_csv('gamma21.700000.csv')
HO = pd.read_csv('gamma24.100000.csv')
O2 = pd.read_csv('gamma150.000000.csv')

plt.plot(x, V(x), 'k', linewidth=1)
plt.hlines(H2.E, 0.9, 2.5, colors=['b', 'g', 'r', 'c', 'm'], linewidths=1)
plt.xlabel('x', fontsize=14)
plt.ylabel('E', fontsize=14, rotation='horizontal')
plt.grid()
pp.savefig()
plt.show()

plt.plot(x, V(x), 'k', linewidth=1)
plt.hlines(HO.E, 0.9, 2.5, colors=['b', 'g', 'r', 'c', 'm'], linewidths=1)
plt.xlabel('x', fontsize=14)
plt.ylabel('E', fontsize=14, rotation='horizontal')
plt.grid()
pp.savefig()
plt.show()

colors = [['b', 'g', 'r', 'c', 'm', 'y'][i%6] for i in range(38)]
plt.plot(x, V(x), 'k', linewidth=1)
plt.xlabel('x', fontsize=14)
plt.ylabel('E', fontsize=14, rotation='horizontal')
plt.hlines(O2.E, 0.9, 2.5, linewidths=1, colors=colors)
plt.grid()
pp.savefig()
plt.show()
```

```python
pp.close()
```
